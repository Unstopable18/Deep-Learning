## Batch Normalization

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
import tensorflow
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense,BatchNormalization
from mlxtend.plotting import plot_decision_regions
import matplotlib.gridspec as gridspec
import itertools
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/Unstopable18/Datasets/main/concertriccir2.csv")
df.head()
```

```python

plt.scatter(df['col-0'],df['col-1'],c=df['col-2'])
plt.show()
```

```python
X=df.iloc[:,0:2].values
y=df['col-2'].values
```

```python
X.shape
```

***

#### tanh

```python
model_tanh = Sequential()

model_tanh.add(Dense(12,activation='tanh',input_dim=2))
model_tanh.add(Dense(8,activation='tanh'))
model_tanh.add(Dense(4,activation='tanh'))
model_tanh.add(Dense(1,activation='sigmoid'))

print(model_tanh.summary())
model_tanh.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
history_tanh=model_tanh.fit(X,y,epochs=100,validation_split=0.2,verbose=0)
plot_decision_regions(X,y.astype(np.int_), clf=model_tanh, legend=2)
```

```python
model_tanh_BN = Sequential()

model_tanh_BN.add(Dense(12,activation='tanh',input_dim=2))
model_tanh_BN.add(BatchNormalization())
model_tanh_BN.add(Dense(8,activation='tanh'))
model_tanh_BN.add(BatchNormalization())
model_tanh_BN.add(Dense(4,activation='tanh'))
model_tanh_BN.add(BatchNormalization())
model_tanh_BN.add(Dense(1,activation='sigmoid'))

print(model_tanh_BN.summary())
model_tanh_BN.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
history_tanh_BN=model_tanh_BN.fit(X,y,epochs=100,validation_split=0.2,verbose=0)
plot_decision_regions(X,y.astype(np.int_), clf=model_tanh_BN, legend=2)
```

***

#### Sigmoid

```python
model_sigmoid = Sequential()

model_sigmoid.add(Dense(12,activation='sigmoid',input_dim=2))
model_sigmoid.add(Dense(8,activation='sigmoid'))
model_sigmoid.add(Dense(4,activation='sigmoid'))
model_sigmoid.add(Dense(1,activation='sigmoid'))

print(model_sigmoid.summary())
model_sigmoid.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
history_sigmoid=model_sigmoid.fit(X,y,epochs=100,validation_split=0.2,verbose=0)
plot_decision_regions(X,y.astype(np.int_), clf=model_sigmoid, legend=2)
```

```python
model_sigmoid_BN = Sequential()

model_sigmoid_BN.add(Dense(12,activation='sigmoid',input_dim=2))
model_sigmoid_BN.add(BatchNormalization())
model_sigmoid_BN.add(Dense(8,activation='sigmoid'))
model_sigmoid_BN.add(BatchNormalization())
model_sigmoid_BN.add(Dense(4,activation='sigmoid'))
model_sigmoid_BN.add(BatchNormalization())
model_sigmoid_BN.add(Dense(1,activation='sigmoid'))

print(model_sigmoid_BN.summary())
model_sigmoid_BN.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
history_sigmoid_BN=model_sigmoid_BN.fit(X,y,epochs=100,validation_split=0.2,verbose=0)
plot_decision_regions(X,y.astype(np.int_), clf=model_sigmoid_BN, legend=2)
```

***

#### ReLU

```python
model_relu = Sequential()

model_relu.add(Dense(12,activation='relu',input_dim=2))
model_relu.add(Dense(8,activation='relu'))
model_relu.add(Dense(4,activation='relu'))
model_relu.add(Dense(1,activation='sigmoid'))

print(model_relu.summary())
model_relu.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
history_relu=model_relu.fit(X,y,epochs=100,validation_split=0.2,verbose=0)
plot_decision_regions(X,y.astype(np.int_), clf=model_relu, legend=2)
```

```python
model_relu_BN = Sequential()

model_relu_BN.add(Dense(12,activation='relu',input_dim=2))
model_relu_BN.add(BatchNormalization())
model_relu_BN.add(Dense(8,activation='relu'))
model_relu_BN.add(BatchNormalization())
model_relu_BN.add(Dense(4,activation='relu'))
model_relu_BN.add(BatchNormalization())
model_relu_BN.add(Dense(1,activation='sigmoid'))

print(model_relu_BN.summary())
model_relu_BN.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
history_relu_BN=model_relu_BN.fit(X,y,epochs=100,validation_split=0.2,verbose=0)
plot_decision_regions(X,y.astype(np.int_), clf=model_relu_BN, legend=2)
```

## Xavier-Glorot for tanH/Sigmoid

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
import tensorflow
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
from mlxtend.plotting import plot_decision_regions
import matplotlib.gridspec as gridspec
import itertools
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/Unstopable18/Datasets/main/ushape.csv")
df.head()
```

```python

plt.scatter(df['col-0'],df['col-1'],c=df['col-2'])
plt.show()
```

```python
X=df.iloc[:,0:2].values
y=df['col-2'].values
```

```python
X.shape
```

#### Glorot Normal

```python
model_tanh_GN = Sequential()

model_tanh_GN.add(Dense(10,activation='tanh',input_dim=2,kernel_initializer='glorot_normal'))
model_tanh_GN.add(Dense(10,activation='tanh',kernel_initializer='glorot_normal'))
model_tanh_GN.add(Dense(10,activation='tanh',kernel_initializer='glorot_normal'))
model_tanh_GN.add(Dense(10,activation='tanh',kernel_initializer='glorot_normal'))
model_tanh_GN.add(Dense(1,activation='sigmoid'))

print(model_tanh_GN.summary())
model_tanh_GN.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
model_tanh_GN.fit(X,y,epochs=100,validation_split=0.2,verbose=0)
plot_decision_regions(X,y.astype(np.int_), clf=model_tanh_GN, legend=2)
```

```python
model_sigmoid_GN = Sequential()

model_sigmoid_GN.add(Dense(10,activation='sigmoid',input_dim=2,kernel_initializer='glorot_normal'))
model_sigmoid_GN.add(Dense(10,activation='sigmoid',kernel_initializer='glorot_normal'))
model_sigmoid_GN.add(Dense(10,activation='sigmoid',kernel_initializer='glorot_normal'))
model_sigmoid_GN.add(Dense(10,activation='sigmoid',kernel_initializer='glorot_normal'))
model_sigmoid_GN.add(Dense(1,activation='sigmoid'))

print(model_sigmoid_GN.summary())
model_sigmoid_GN.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
model_sigmoid_GN.fit(X,y,epochs=100,validation_split=0.2,verbose=0)
plot_decision_regions(X,y.astype(np.int_), clf=model_sigmoid_GN, legend=2)
```

#### He Normal

```python
model_relu_HN = Sequential()

model_relu_HN.add(Dense(10,activation='relu',input_dim=2,kernel_initializer='he_normal'))
model_relu_HN.add(Dense(10,activation='relu',kernel_initializer='he_normal'))
model_relu_HN.add(Dense(10,activation='relu',kernel_initializer='he_normal'))
model_relu_HN.add(Dense(10,activation='relu',kernel_initializer='he_normal'))
model_relu_HN.add(Dense(1,activation='sigmoid'))

print(model_relu_HN.summary())
model_relu_HN.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
model_relu_HN.fit(X,y,epochs=100,validation_split=0.2,verbose=0)
plot_decision_regions(X,y.astype(np.int_), clf=model_relu_HN, legend=2)
```

# Plot Comparison

```python
gs = gridspec.GridSpec(1, 2)

fig = plt.figure(figsize=(12,4))
ylabels = ['Relu_He', 'Relu_He']
labels = ['Normal', 'Uniform']
for clf, lab,ylab, grd in zip([model_relu_HN, model_relu_HU],labels,ylabels,itertools.product([0, 1], repeat=2)):
    ax = plt.subplot(gs[grd[0], grd[1]])
    fig = plot_decision_regions(X,y.astype(np.int_), clf=clf, legend=2)
    plt.title(lab)
    plt.ylabel(ylab)
plt.show()
```

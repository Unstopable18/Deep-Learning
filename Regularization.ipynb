```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
import seaborn as sns
from sklearn.model_selection import train_test_split
from keras import Sequential
from keras.layers import Dense,Dropout
from keras.regularizers import l2,l1
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.datasets import make_moons
```

```python
X,y=make_moons(n_samples=500, noise=0.05, random_state=42)
plt.scatter(X[:,0,],X[:,1],c=y,s=100)
plt.show()
```

```python
X.shape
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
model = Sequential()
model.add(Dense(128,activation='relu',kernel_regularizer=l2(0.03),input_dim=2))
model.add(Dense(64,activation='relu',kernel_regularizer=l2(0.03)))
model.add(Dense(1,activation='sigmoid'))
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
callback = EarlyStopping(monitor="accuracy",min_delta=0.00001,patience=50,verbose=1,mode="auto",baseline=None,restore_best_weights=False)
print(model.summary())
regular_callback = model.fit(X_train,y_train,validation_data=(X_test,y_test),batch_size=455,epochs=100,callbacks=callback)
```

```python
model = Sequential()
model.add(Dense(128,activation='relu',kernel_regularizer=l2(0.03),input_dim=2))
model.add(Dropout(0.2))
model.add(Dense(64,activation='relu',kernel_regularizer=l2(0.03)))
model.add(Dense(1,activation='sigmoid'))
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
callback = EarlyStopping(monitor="accuracy",min_delta=0.00001,patience=50,verbose=1,mode="auto",baseline=None,restore_best_weights=False)
print(model.summary())
regular_dropout_callback = model.fit(X_train,y_train,validation_data=(X_test,y_test),batch_size=455,epochs=100,callbacks=callback)
```

```python
model = Sequential()
model.add(Dense(128,activation='relu',kernel_regularizer=l2(0.03),input_dim=2))
model.add(Dense(64,activation='relu',kernel_regularizer=l2(0.03)))
model.add(Dense(1,activation='sigmoid'))
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
print(model.summary())
regular = model.fit(X_train,y_train,validation_data=(X_test,y_test),batch_size=455,epochs=100)
```

```python

f=plt.figure()
f.set_figheight(5)
f.set_figwidth(12)

plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'],label='Accuracy',linestyle='--')
plt.plot(history.history['val_accuracy'],label='Val_Accuracy')
plt.ylabel('Model')
plt.xlabel('Accuracy')

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'],label='loss',linestyle='--')
plt.plot(history.history['val_loss'],label='Val_loss')
plt.xlabel('Loss')

plt.legend()
```
